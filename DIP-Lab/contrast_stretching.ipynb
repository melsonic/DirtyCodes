`Contrast Stretching`

```python
import cv2
import os
from matplotlib import pyplot as plt

try:
  image = cv2.imread("img.jpeg")
except IOError:
  pass

""" rgb_bands = img.split()
red, green, blue = rgb_bands[0], rgb_bands[1], rgb_bands[2]
rmin, rmax = 40, 200
gmin, gmax = 50, 180
bmin, bmax = 60, 170

def get_normalized_red(intensity):
  norm_int = (intensity - rmin)*(255/(rmax-rmin))
  return norm_int

def get_normalized_green(intensity):
  norm_int = (intensity - gmin)*(255/(gmax-gmin))
  return norm_int

def get_normalized_blue(intensity):
  norm_int = (intensity - bmin)*(255/(bmax-bmin))
  return norm_int

r_band = rgb_bands[0].point(get_normalized_red)
g_band = rgb_bands[1].point(get_normalized_green)
b_band = rgb_bands[2].point(get_normalized_blue)

contrast_stret = Image.merge("RGB", (r_band, g_band, b_band)) """

stretched_image = image.copy()
# get height and width of the image 
height, width, _ = image.shape 

maxiI = 250
miniI = 3

maxoI = image.max()
minoI = image.min()
  
for i in range(0, height - 1): 
    for j in range(0, width - 1): 
          
        # Get the pixel value 
        pixel = stretched_image[i, j] 
          
        # scale each pixel by this formula
        '''
        pout = (pin - miniI) * ((maxoI-minoI) / (maxiI-miniI)) + minoI 
        
        '''
        
          
        # 1st index contains red pixel 
        pixel[0] = (pixel[0] - miniI) * ((maxoI-minoI) / (maxiI-miniI)) + minoI  
          
        # 2nd index contains green pixel 
        pixel[1] = (pixel[1] - miniI) * ((maxoI-minoI) / (maxiI-miniI)) + minoI 
          
        # 3rd index contains blue pixel 
        pixel[2] = (pixel[2] - miniI) * ((maxoI-minoI) / (maxiI-miniI)) + minoI 
          
        # Store new values in the pixel 
        stretched_image[i, j] = pixel 
```
