`Histogram Stretching`

```python
from PIL import Image, ImageColor
import os
from matplotlib import pyplot as plt

try:
  imgpath = os.getcwd() + "/img.jpeg"
  img = Image.open(imgpath)
except IOError:
  pass

rgb_bands = img.split()
red, green, blue = rgb_bands[0], rgb_bands[1], rgb_bands[2]
rmin, rmax = 40, 200
gmin, gmax = 50, 180
bmin, bmax = 60, 170

def get_normalized_red(intensity):
  norm_int = (intensity - rmin)*(255/(rmax-rmin))
  return norm_int

def get_normalized_green(intensity):
  norm_int = (intensity - gmin)*(255/(gmax-gmin))
  return norm_int

def get_normalized_blue(intensity):
  norm_int = (intensity - bmin)*(255/(bmax-bmin))
  return norm_int

r_band = rgb_bands[0].point(get_normalized_red)
g_band = rgb_bands[1].point(get_normalized_green)
b_band = rgb_bands[2].point(get_normalized_blue)

hist_stretching = Image.merge("RGB", (r_band, g_band, b_band))
```

#### Before Histogram Stretching

```python
plt.imshow(img)
```

#### After Histogram Stretching

```python
plt.imshow(hist_stretching)
```
