```python
import cv2 as cv
from PIL import Image
import matplotlib.pyplot as plt
import os

img = cv.imread('img.jpeg')
```

#### Before Histogram Equqlization

```python
plt.imshow(img)
```

```python
R, G, B = cv.split(img)
img2_R = cv.equalizeHist(R)
img2_G = cv.equalizeHist(G)
img2_B = cv.equalizeHist(B)
img2 = cv.merge((img2_R, img2_G, img2_B))
```

#### After Histogram Equalization

```python
plt.imshow(img2)
```
